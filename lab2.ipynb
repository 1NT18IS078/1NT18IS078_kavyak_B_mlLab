```python
#standard deviation and variation
import math
list1 = [1,2,3,4]
temp = 0
mean = sum(list1)/len(list1)
for i in list1:
    temp = temp+((i-mean)**2)
sd = math.sqrt((temp/(len(list1))))
var=sd*sd
print("standard deviation is : {}".format(sd))
print("variance is :{}".format(var))
```

```python
#min-max normalization
a=(115.3,195.5,120.5,110.2,90.4,105.6,110.9,116.3,122.3,125.4)
xmin= min(a)
xmax= max(a)
print("value: normalized value")
for x in a:
    n= (x-xmin)/(xmax-xmin)
    print("{} : {}" .format(x,n))
    
```

```python
#standardization
import math
temp_sum = 0
a = (115.3,195.5,120.5,110.2,90.4,105.6,110.9,116.3,122.3,125.4)
mean = sum(a)/len(a)
for i in a:
    temp_sum = temp_sum + ((i-mean)**2)
sd = math.sqrt((temp_sum/(len(a))))
print("value : standardized value")
for i in a:
    s = (i-mean)/sd
    print("{} : {}" .format(i,s))
        
```
